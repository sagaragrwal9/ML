```python
import pandas as pd 
import numpy as np 
```

```python
df=pd.read_csv(r'C:\Users\sagar\Desktop\HR_comma_sep.csv')
```

```python
df
```

```python
x = df[['satisfaction_level','average_montly_hours','promotion_last_5years','salary']]
x
```

```python
salary_dummies = pd.get_dummies(x.salary, prefix="salary")
```

```python
df_with_dummies = pd.concat([x,salary_dummies],axis='columns')
df_with_dummies.drop('salary',axis='columns',inplace=True)
df_with_dummies.head()
y=df.left
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_with_dummies,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
X_test
```

```python

model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
