```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_excel('can.xlsx')
df
```

```python

df.rename(columns={'per capita income (US$)':'income'},inplace=True)
df.dropna()
```

```python

%matplotlib inline
plt.xlabel("year")
plt.ylabel("per capita income")
plt.scatter(df.year,df.income,marker='+',color='red')
df.dropna()
```

```python
reg=linear_model.LinearRegression()
year=df.drop('income',axis='columns')
income=df.income
```

```python
year=year.dropna()
income=income.dropna()
```

```python
reg.fit(year,income)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
